```python
# 1. Create a class
class ElectricVehicle:
    
    # 2. Declare attributes in __init__
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        self.battery = 100  # Start with 100% battery
    
    # 3. Create a method to display information
    def show_info(self):
        print(f"{self.brand} {self.model} - Battery: {self.battery}%")
    
    # Another method to process information
    def drive(self):
        if self.battery > 0:
            self.battery -= 20  # Driving uses 20% battery
            print(f"Drove {self.brand} {self.model}. Battery now: {self.battery}%")
        else:
            print("Battery empty! Please charge.")

# 4. Create objects (instantiate)
ev1 = ElectricVehicle("Tesla", "Model 3")
ev2 = ElectricVehicle("Nissan", "Leaf")

# 5. Access attributes and methods using objects
print("=== Initial Status ===")
ev1.show_info()
ev2.show_info()

print("\n=== Driving ===")
ev1.drive()
ev1.drive()
ev2.drive()

print("\n=== Final Status ===")
ev1.show_info()
ev2.show_info()

# Access attributes directly
print(f"\n=== Direct Attribute Access ===")
print(f"EV1 is a {ev1.brand}")
print(f"EV2 is a {ev2.model}")
```
