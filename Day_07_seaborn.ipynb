```python
data = { "Year":[2019, 2020, 2021, 2022, 2023],
         "Sales_Units":[5000, 7000, 12000, 20000, 35000],
         "Revenue_Million":[50, 75, 140, 250, 500],
         "Profit_Margin":[5, 6, 8, 10,15],
         "Charging_Stations":[200,400,800,1500,3000]
       }
data
```

```python
import seaborn as sns
ev_df = pd.DataFrame(data)
```

```python
ev_df
```

```python
sns.set(style="whitegrid")
```

```python
sns.lineplot(x="Year", y="Sales_Units", data=ev_df, marker='o')
plt.title("EV Sales Growth Trend")
plt.show()
```

```python
ev_df.columns
```

```python
sns.scatterplot(x="Charging_Stations", y="Sales_Units", data=ev_df)
plt.title("Charging stations vs Sales")
plt.show()
```

```python
sns.barplot(x="Year", y="Revenue_Million", data=ev_df)
plt.title("Revenue Comparision")
plt.show()
```

```python
sns.boxplot(y="Profit_Margin", data=ev_df)
plt.title("Profit Margin Distribution")
plt.show()
```

```python
sns.histplot(ev_df["Sales_Units"], kde=True)#Kernel Density Estimation
plt.title("Sales Distribution")
plt.show()
```

```python
corr = ev_df.corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(ev_df)
plt.show()
```

```python
sns.violinplot(y="Sales_Units",data=ev_df)
plt.title("Sales Distribution Shapes")
plt.show()
```

```python
ev_df["Region"]=["North", "South", "East", "West", "North"]

sns.countplot(x="Region", data=ev_df)
plt.title("Region-wise Records")
plt.show()
```

```python
sns.regplot(x="Charging_Stations", y="Sales_Units",data=ev_df)
plt.title("Regression Analysis")
plt.show()
```
