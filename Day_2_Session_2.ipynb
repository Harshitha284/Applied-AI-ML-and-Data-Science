### Identity Operators

```python
a = ['a', 'b', 'c']
b = ['a', 'b', 'c']
print('a is b' , a is b)
```

```python
a = ['a', 'b', 'c']
b =  a
print('a is b' , a is b)
```

```python
id(a) == id(b)
```

```python
a = ['a', 'b', 'c']
b = ['a', 'b', 'c']
print('a is not b' , a is not b)
```
