```python
#Scenario: EV fleet energy & Cost Management System
#Everyday company needs to
#----------------------------------
#Calculate Energy consumed 
#Estmated charging cost
#Apply Dynamic electricity pricing
#Generate daily report
#Handle different vehicles with different parameters
```

```python
def calculate_energy(distance_km, efficiency_km_per_kwh):
    energy_used = distance_km / efficiency_km_per_kwh
    return energy_used
```

```python
def calculate_charging_cost(energy_kwh, price_per_kwh=8):
    cost = energy_kwh * price_per_kwh
    return cost
```

```python
def total_fleet_energy(*energy_values):
    total_energy = sum(energy_values)
    return total_energy
```

```python
def vehicle_summary(**vehicle_info):
    for key, value in vehicle_info.items():
        print(key, ":", value)
```

```python
def battery_drain_simulator(start_charge):
    while start_charge > 0:
        yield start_charge
        start_charge -= 10
```

```python
energy_car1 = calculate_energy(120, 6)
energy_car2 = calculate_energy(90, 5)

cost_car1 = calculate_charging_cost(energy_car1)
cost_car2 = calculate_charging_cost(energy_car2, 9)

fleet_energy = total_fleet_energy(energy_car1, energy_car2)
print("Energy car 1:", energy_car1)
print("Energy car 2:", energy_car2)

print("Charging cost car 1:", cost_car1)
print("Charging cost car 2:", cost_car2)

print("Total fleet Energy:", fleet_energy)

vehicle_summary(Vehicle_Id = "EV1023", city="Bangalore", Driver = "Rahul", Battery ="Lithium-Ion")

for level in battery_drain_simulator(100):
    print("Battery Level:", level)
```
