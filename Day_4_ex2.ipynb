```python
# Parent class with different access levels
class Vehicle:
    def __init__(self, brand):
        self.brand = brand              # Public
        self._engine_type = "Electric"  # Protected
        self.__vin = "ABC123"           # Private
    
    def show_public(self):
        return f"Brand: {self.brand}"
    
    def _show_protected(self):
        return f"Engine: {self._engine_type}"
    
    def __show_private(self):
        return f"VIN: {self.__vin}"


# Child class
class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model
    
    def test_access(self):
        print("✓ Public:", self.show_public())    # Works
        print("✓ Protected:", self._show_protected())  # Works
        # print("✗ Private:", self.__show_private())  # Error!


# Testing
car = Car("Tesla", "Model 3")
car.test_access()

print("\nDirect access from outside:")
print("✓ Public attr:", car.brand)
print("✓ Protected attr (not recommended):", car._engine_type)
# print("✗ Private attr:", car.__vin)  # Error!
```
