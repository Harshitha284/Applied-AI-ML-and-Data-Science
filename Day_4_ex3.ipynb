```python
class ev:
    def charge(self):
        return "Charging at standard speep"
    def drive(self):
        return "Driving electric vehicle"

class telsa(ev):
    def charge(self):
        return "Super charging at 250KW"
    def drive(self):
        return "Driving with autopilot"

class Nissan(ev):
    def charge(self):
        return "Charging at 50KW"

ev1=ev()
ev2=telsa()
ev3=Nissan()

print("Polimorphism in action")
for vehicle in [ev1, ev2, ev3]:
    print(f"-{vehicle.charge()}")
```

```python
#Method overloading

class Calculator:
    def add(self, a, b=None, c=None):
        if b is None and c is None:
            return a
        elif c is None:
            return a+b
        else:
            return a+b+c

calc= Calculator()

print("Method overloading simulation")
print(f"add(5)={calc.add(5)}")
print(f"add(5, 3)={calc.add(5, 3)}")
print(f"add(5, 3, 2)={calc.add(5, 3, 2)}")           
```

```python
#Method Overriding
class Payment:
    def process_payment(self, amount):
        return f"Processing ${amount} with generic payment"

class CreditCardPayment(Payment):
    def process_payment(self, amount):
        return f"Processing ${amount} via Credit Card"

class PayPalPayment(Payment):
    def process_payment(self, amount):
        return f"Processing ${amount} via PayPal"

# Usage
payments = [Payment(), CreditCardPayment(), PayPalPayment()]
amount = 100

for payment in payments:
    print(payment.process_payment(amount))
```

```python
# PARENT CLASS
class Animal:
    def __init__(self, name):
        self.name = name
    
    def make_sound(self):
        return "Some generic animal sound"
    
    def move(self):
        return f"{self.name} is moving"

# CHILD CLASS 1
class Dog(Animal):
    # Constructor overriding (extending)
    def __init__(self, name, breed):
        super().__init__(name)  # Call parent constructor
        self.breed = breed
    
    # Method overriding
    def make_sound(self):
        return "Woof! Woof!"
    
    # Additional method (not overriding)
    def wag_tail(self):
        return f"{self.name} is wagging tail"

# CHILD CLASS 2  
class Cat(Animal):
    def make_sound(self):
        return "Meow!"

# Test
print("=== Method Overriding Demo ===")

# Create objects
animal = Animal("Generic Animal")
dog = Dog("Buddy", "Golden Retriever")
cat = Cat("Whiskers")

# Same method, different behaviors
print(f"\n1. Animal sound:")
print(f"   {animal.name}: {animal.make_sound()}")
print(f"   {dog.name}: {dog.make_sound()}")
print(f"   {cat.name}: {cat.make_sound()}")

print(f"\n2. Movement (inherited, not overridden):")
print(f"   {dog.move()}")  # Uses parent's method
print(f"   {cat.move()}")  # Uses parent's method

print(f"\n3. Dog's special method:")
print(f"   {dog.wag_tail()}")

# Type checking
print(f"\n4. Type Relationships:")
print(f"   Is dog an Animal? {isinstance(dog, Animal)}")  # True
print(f"   Is dog a Dog? {isinstance(dog, Dog)}")        # True
```
