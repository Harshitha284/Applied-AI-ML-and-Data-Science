```python
data = { "year":[2019, 2020, 2021, 2022, 2023],
         "Sales_Units":[5000, 7000, 12000, 20000, 35000],
         "Revenue_Million":[50, 75, 140, 250, 500],
         "Profit_Margin":[5, 6, 8, 10,15]
       }
data
```

```python
ef_df = pd.DataFrame(data)
```

```python
ef_df
```

```python
plt.plot((ef_df["year"], ef_df["Sales_Units"]))
```

```python
plt.plot((ef_df["year"], ef_df["Sales_Units"]))
plt.title("EV Sales Growth Over Year")
plt.xlabel('Year')
plt.ylabel("Sales Unit")
plt.show()
          
```

```python
plt.scatter(ef_df['Sales_Units'],ef_df['Revenue_Million'])
plt.title("Sales vs Revenue Relationship")
plt.xlabel('Sales Unit')
plt.ylabel("Revenue(Million")
plt.show()
```

```python
plt.bar(ef_df["year"], ef_df["Revenue_Million"])
plt.title("Yearwise Revenue Comparision")
plt.xlabel('Year')
plt.ylabel("Revenue(Million)")
plt.show()
```

```python
plt.boxplot(ef_df["Profit_Margin"])
plt.title("Profit Margin Distribution")
plt.ylabel('Profit Margin (%)')
plt.show()
```

```python
import numpy as np
import pandas as pd
```

```python
categories = ['Sales', 'Revenue', 'Profit' 'Stations']
values = [35000, 500, 15, 3000]

angles = np.linspace(0, 2*np.pi, len(categories), endpoint=False)
values = np.concatenate((values, [values[0]]))
angles = np.concatenate((angles, [angles[0]]))

fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, values)
ax.fill(angles, values, alpha=0.25)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
plt.title("EV Comapany Performace")
plt.show()
```

```python

from mpl_toolkits.mplot3d import Axes3D
data = { "year":[2019, 2020, 2021, 2022, 2023],
         "Sales_Units":[5000, 7000, 12000, 20000, 35000],
         "Revenue_Million":[50, 75, 140, 250, 500],
         "Profit_Margin":[5, 6, 8, 10,15]
       }
data
ef_df = pd.DataFrame(data)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(ef_df['year'],ef_df['Sales_Units'],ef_df['Revenue_Million'])

ax.set_xlabel("Year")
ax.set_ylabel("Sales_Units")
ax.set_zlabel("Revenue")

plt.title("3D EV Business Analysis")
plt.show()
```
